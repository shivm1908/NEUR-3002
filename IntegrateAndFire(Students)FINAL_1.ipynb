```python
# Jupyter: enable plot rendering inside the notebook (this allows us to actually show the output of the code visually)
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

# --- Neuron Parameters (realistic, in mV and pA) ---
C = 300.0  # this is the value for capacitance
V_th = 1 -55.0 #threshold needed for volatge in order for sodium channels to open
V_reset = -65.0 # value after the neuron resets
V_rest = -65.0 #volatge with no current, in this case the values would be the same since there is no decay
I_ext = 600.0 # current
dt = 0.1


T = 100          # total simulation time in ms
dt = 0.1         # time step in ms
time = np.arange(0, T, dt)  # time vector

# --- Initialize voltage and spike times ---
V = np.full_like(time, V_rest)  # set all values to resting voltage
spike_times = []

# --- Simulation Loop ---
for i in range(1, len(time)):
    dV = (I_ext / C) * dt       # V = (I / C) * dt, in mV
    V[i] = V[i-1] + dV

    if V[i] >= V_th:
        V[i-1] = 40.0           # mark spike as +40 mV (for visualization)
        V[i] = V_reset
        spike_times.append(time[i])

# --- Plotting ---
plt.figure(figsize=(10, 4))
plt.plot(time, V, label='Voltage')
plt.axhline(V_th, color='r', linestyle='--', label='Threshold')
plt.xlabel('time') #what should the x be?
plt.ylabel('membrane voltage') #what should the y be?
plt.title('Integrate and Fire Models') #what is a good title?
plt.legend() #can you think of a legend?
plt.grid(True)
plt.show()

```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

# --- Neuron Parameters ---
C = 300.0 # this is the value for capacitance
V_th #threshold needed for volatge in order for sodium channels to open
V_reset = -65.0 # value after the neuron resets
V_rest = -65.0 #volatge with no current, in this case the values would be the same since there is no decay
dt = 0.1

T = 100
time = np.arange(0, T, dt)

# --- Different input currents (different neurons/conditions) ---
input_currents = [300, 500, 700]  # multiple conditions

plt.figure(figsize=(10, 4))

for I_ext in input_currents:
    V = np.full_like(time, V_rest)

    for i in range(1, len(time)):
        dV = (I_ext / C) * dt
        V[i] = V[i-1] + dV

        if V[i] >= V_th:
            V[i-1] = 40.0   # spike for visualization
            V[i] = V_reset

    plt.plot(time, V, label=f'I = {I_ext} pA')

# --- Plot formatting ---
plt.axhline(V_th, linestyle='--', label='Threshold')
plt.xlabel('Time (ms)')
plt.ylabel('Membrane Voltage (mV)')
plt.title('Integrate-and-Fire Model with Different Input Currents')
plt.legend()
plt.grid(True)
plt.show()

```
